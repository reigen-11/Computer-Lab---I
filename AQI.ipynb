```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AQI.csv", low_memory=False)
```

```python
num_cols = df.select_dtypes(include="number").columns.tolist()
df = df.fillna(df.mean(numeric_only=True))
```

```python
df = df.dropna()
```

```python
df['sampling_date'] = pd.to_datetime(df['sampling_date'], format='%d-%m-%Y', errors='coerce')
```

```python
plt.scatter(df['sampling_date'], df['so2'], label='SO2', color='blue', alpha=0.7)
plt.scatter(df['sampling_date'], df['no2'], label='NO2', color='green', alpha=0.7)
plt.scatter(df['sampling_date'], df['rspm'], label='RSPM', color='orange', alpha=0.7)
plt.scatter(df['sampling_date'], df['spm'], label='SPM', color='red', alpha=0.7)
plt.scatter(df['sampling_date'], df['pm2_5'], label='PM2.5', color='purple', alpha=0.7)

plt.title('Pollutant Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Concentration')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
